```python
mb_size = [0.1, 1, 2, 5, 10, 20, 40, 50, 80, 100, 120, 150, 160]
data = [2.61859893799, 14.0395522118, 21.5435862541, 45.8077788353, 86.766576767, 166.588747501, 367.716157436, 466.95634126, 704.421913624, 916.277551651, 1032.60505199, 1343.70471239, 1422.89692163]

fig = plt.figure()
SIZE_MB = [0.001 * pow(1.23, i) for i in range(4)] # exponential data_size growth
x = mb_size
plt.plot(x, np.asarray(data))
plt.title("ROS Communication Overhead")
plt.xlabel("Message Size: (MB)")
plt.ylabel("Overhead: (ms)")
plt.legend()
plt.show()
fig.savefig("ros_time.png")
```

```python
throughputs = []
for i in range(len(data)):
    throughputs.append(data[i] / 1000 / (mb_size[i] / 1000))

fig = plt.figure()
# SIZE_MB = [0.001 * pow(1.23, i) for i in range(4)] # exponential data_size growth
x = mb_size
plt.plot(x, np.asarray(throughputs))
plt.title("ROS Communication Throughput")
plt.xlabel("Message Size: (MB)")
plt.ylabel("Throughput: (GB/s)")
plt.legend()
plt.show()
fig.savefig("ros_throughput.png")
```
